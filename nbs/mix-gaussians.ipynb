# mix-gaussians

> Several VAE examples use mixtures of guassian distributions. This notebook explores them.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import torch
```

```python
# Gaussian Parameters
torch.manual_seed(1442)
n_guassians = 25
mu = torch.randn(n_guassians)
sigma = torch.randn(n_guassians) ** 2 # This should probably have different params
weights = torch.rand(n_guassians)
weights /= weights.sum()
```

```python
def sample():
    return torch.normal(mu, sigma)[torch.multinomial(weights, 1).item()]
```

```python
n_samples = 10000
samples = torch.tensor([sample() for _ in range(n_samples)])
```

```python
_ = sns.kdeplot(samples.numpy())
```
